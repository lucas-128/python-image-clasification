```python
a = silkroad_items_short[~silkroad_items_short.category.str.contains("Alcohol")]
```

```python
a = None
silkroad_items_short = None
```

```python
data
```

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
ids = data['image_id']
```

```python
data.dropna(inplace = True)
data.image_id = data.image_id.astype(int)
```

```python
import os.path
```

```python
dict = {}

for id in ids:
  ruta = "images_by_id/" + str(id) + ".jpg"
  if os.path.exists(ruta):
    if id in dict.keys():
      im = dict[id]
    else:
      im = Image.open(ruta)
    if(np.array(im).shape == (135,180,3)):
      dict[id] = np.array(im)
```

```python
valores = pd.DataFrame(list(dict.items()),columns = ['image_id','values'])
```

```python
dataset = valores.merge(data,on='image_id')
```

```python
valores = None
data = None
```

```python
dataset.drop('image_id', axis=1, inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset['values'], dataset['category'], test_size=0.2, random_state=5)
```

```python
X_train[0].shape
```

```python
y_train
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
le.fit(y_train)
```

```python
y_train = le.transform(y_train)
```

```python
y_train[:100]
```

```python
eye = np.eye(max(y_train.flatten())+1)
y_train_labels = np.asarray([eye[i] for i in y_train.flatten()])
y_train_labels
```

```python
y_train_labels[:20]
```

```python
y_train_labels.shape
```

```python
#!pip install tensorflow
```

```python
from tensorflow.keras.layers import Dense, Input, Conv2D, GlobalMaxPooling2D, Concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Model
```

```python
entrada = Input(shape=(135,180,1))

cnn1 = Conv2D(30, (4,4),1)(entrada)
cnn2 = Conv2D(30, (2,2),1)(cnn1)
cnn3 = Conv2D(30, (1,1),1)(cnn2)

result = GlobalMaxPooling2D()(cnn3)

capa1 = Dense(40, activation="relu")(result)
capa2 = Dense(30, activation="relu")(capa1)
capa3 = Dense(20, activation="relu")(capa2)

salida = Dense(10, activation="softmax")(capa3)

model = Model(inputs=entrada, outputs=salida)
```

```python
from tqdm import tqdm
```

```python
def preprocess_image(image):
    return np.dot(image, [1,1,1])/np.sqrt(255)
```

```python
train_images = np.asarray([preprocess_image(img) for img in tqdm(X_train, position=0)])
```

```python
train_images.shape
```

```python
model.compile(optimizer=Adam(0.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
y_train_labels.shape
```

```python
hist = model.fit(train_images.reshape(((11420,135,180,1))), y_train_labels, batch_size=32, epochs=100, validation_split=0.1)
```

```python
y_test_labels = le.transform(y_test)
```

```python
test_images = np.asarray([preprocess_image(img) for img in tqdm(X_test, position=0)])
```

```python
preds = model.predict(test_images)
```

```python
preds = np.argmax(preds, axis=1)
```

```python
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, ConfusionMatrixDisplay, confusion_matrix
```

```python
accuracy_score(y_test_labels.flatten(), preds)
```

```python
print(classification_report(y_test_labels.flatten(), preds))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(dpi = 150)

plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```

```python
plt.figure(dpi = 150)


plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```

```python
import sklearn.metrics as metrics
```

```python
confusion_matrix = metrics.confusion_matrix(y_true=y_test_labels, y_pred=preds,normalize = 'true')
```

```python
plt.figure(dpi = 150)
ax = sns.heatmap(confusion_matrix,linewidths=.5,annot = True,cmap = 'Blues')

sns.set(font_scale=0.55)

ax.set_xlabel('Predicción')
ax.set_ylabel('Valor real')
ax.set_title('Matriz de confusión para las predicciones realizadas')

ax.set(xticklabels = labels)
ax.set(yticklabels = labels)

plt.xticks(rotation=90)
plt.yticks(rotation=0)
```
